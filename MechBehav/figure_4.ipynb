```python
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np


sb.set_theme(style="ticks", context="paper")

plt.rcParams["text.usetex"] = True

fig, axs = plt.subplots(2, 1, figsize=(4, 4))

# ising plot
ax = axs[0]
J2s = np.linspace(0.1, 10, 10)
tcs = np.load("ising_results/tcs.npy")
dtcs = np.load("ising_results/dtcs.npy")
# tcs_line = interpolate.make_interp_spline(J2s, tcs)
# dtcs_line = interpolate.make_interp_spline(J2s, dtcs)
# J2s = np.linspace(0.1, 10, 500)
# tcs = tcs_line(J2s)
# dtcs = dtcs_line(J2s)

ax.plot(J2s, tcs, color="#224b87", label=r"$\langle \mathcal{T}_3 \rangle$")
ax.plot(J2s, dtcs, color="#f47900", label=r"$\langle \mathcal{D}_3 \rangle$")
ax.set_xlabel(r"$J_2$")
ax.legend(loc="upper left")

# simplagion plot
ax = axs[1]
lambdas = np.linspace(0.1, 10, 10)
tcs = np.load("simplagion_results/tcs.npy")
dtcs = np.load("simplagion_results/dtcs.npy")
# tcs_line = interpolate.make_interp_spline(lambdas, tcs)
# dtcs_line = interpolate.make_interp_spline(lambdas, dtcs)
# lambdas = np.linspace(0.1, 10, 1000)
# tcs = tcs_line(lambdas)
# dtcs = dtcs_line(lambdas)

ax.plot(lambdas, tcs, color="#224b87", label=r"$\langle \mathcal{T}_3 \rangle$")
ax.plot(lambdas, dtcs, color="#f47900", label=r"$\langle \mathcal{D}_3 \rangle$")
ax.set_xlabel(r"$\lambda_2$")
ax.legend(loc="upper left")

axs[0].annotate(r"$\bf{A}$", (-0.1, 1.1), xycoords="axes fraction")
axs[1].annotate(r"$\bf{B}$", (-0.1, 1.21), xycoords="axes fraction")

sb.despine()
fig.tight_layout()

fig.savefig("./figures/figure4_draft.pdf", dpi=300, bbox_inches="tight")
```
